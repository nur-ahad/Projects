```python
#Stem Plot Creation 
np.random.seed(3)
```

```python
x =0.5+np.arange(8)
y =np.random.uniform(2,7,len(x))
```

```python
fig,ax =plt.subplots()
ax.stem(x,y)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
#Step Plot Creation
np.random.seed(3)
```

```python
x =0.5+np.arange(8)
y =np.random.uniform(2,7,len(x))
```

```python
fig,ax =plt.subplots()
ax.step(x,y)
ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
#Fill_inbetween Graph Creation 
np.random.seed(1)
```

```python
x = np.linspace(0,8,16)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x))
```

```python
fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, alpha=.5, linewidth=0)
ax.plot(x, (y1 + y2)/2, linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, alpha=0.1,linewidth=0)
ax.plot(x, (y1 + y2)/2, linewidth=2)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```
